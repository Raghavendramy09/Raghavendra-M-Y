```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
# load the mtcars dataset
df = pd.read_csv(r"C:\Users\PUNITH KUMAR C\Desktop\AIML study materials\Visualization_programs\mtcars.csv")
```

```python
df.head()
```

```python
# visualize the distribution of the mpg column
plt.hist(df['mpg'])
plt.xlabel('Miles per Gallon')
plt.ylabel('Frequency')
plt.title('Distribution of Miles per Gallon')
plt.show()
```

```python
# visualize the relationship between mpg and wt
plt.scatter(df['wt'], df['mpg'])
plt.xlabel('Weight (lbs)')
plt.ylabel('Miles per Gallon')
plt.title('Relationship between Weight and Miles per Gallon')
plt.show()
```
